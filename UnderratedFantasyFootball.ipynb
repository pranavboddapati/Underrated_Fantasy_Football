```python
import pandas as pd

url = "https://www.pro-football-reference.com/years/2024/fantasy.htm"
df = pd.read_html(url)[0]

print("Columns:")
print(df.columns.tolist())

df.head()
```

```python
df.columns.tolist()
```

```python
#flatten multi-level column names
df.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in df.columns]

print("Flattened columns:")
print(df.columns.tolist())

df.head()
```

```python
#rename important columns
df = df.rename(columns={
    'Unnamed: 1_level_0_Player': 'Player',
    'Unnamed: 2_level_0_Tm': 'Team',
    'Unnamed: 3_level_0_FantPos': 'Position',
    'Fantasy_FantPt': 'FantasyPoints',
    'Games_G': 'GamesPlayed'
})

#drop rows with missing player name or fantasy points
df = df[df["Player"].notna() & df["FantasyPoints"].notna()]

#convert GamesPlayed to numeric before filtering
df["GamesPlayed"] = pd.to_numeric(df["GamesPlayed"], errors="coerce")

#remove players who didn’t play at all
df = df[df["GamesPlayed"] > 0]

#convert all stat columns to numeric
non_numeric = ["Player", "Team", "Position"]
numeric_cols = [col for col in df.columns if col not in non_numeric]
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")

#reset index
df = df.reset_index(drop=True)

print(f"Cleaned dataset has {df.shape[0]} players and {df.shape[1]} columns.")
df.head()
```

```python
df.to_csv("cleaned_2024_fantasy.csv", index=False)
```

```python
#importing sqlite database
from google.colab import drive
import sqlite3
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

drive.mount('/content/drive')
db_path = '/content/drive/MyDrive/DataManagementProject/fantasy.db'
conn = sqlite3.connect(db_path)

#new dataframe
df = pd.read_sql("SELECT * FROM fantasy_2024_stats", conn)
df.head()
```

```python
#qb subset
QB_df = df[df['Position'] == 'QB'].copy()

#fantasy ppg
QB_df['FantasyPointsPerGame'] = QB_df['FantasyPoints'] / QB_df['GamesPlayed']

#rush threat
QB_df['RushThreat'] = QB_df['Rushing_Yds'] / QB_df['Rushing_Att'].replace(0, 0.01)

#int rate
QB_df['INTperAtt'] = QB_df['Passing_Int'] / QB_df['Passing_Att'].replace(0, 0.01)

#underrated score
QB_df['UnderratedScore'] = (
    QB_df['RushThreat'] * 0.3 +
    QB_df['Passing_TD'] * 0.2 +
    QB_df['Passing_Yds'] * 0.1 +
    QB_df['FantasyPointsPerGame'] * 0.3 -
    QB_df['INTperAtt'] * 0.1
)


QB_df.sort_values('UnderratedScore', ascending=False).head(10)
```

```python
#rb subset
RB_df = df[df['Position'] == 'RB'].copy()

#fantasy ppg
RB_df['FantasyPointsPerGame'] = RB_df['FantasyPoints'] / RB_df['GamesPlayed']

#metrics
RB_df['FumblesTotal'] = RB_df['Fumbles_Fmb'] + RB_df['Fumbles_FL']
RB_df['YardsPerRush'] = RB_df['Rushing_Yds'] / RB_df['Rushing_Att'].replace(0, 0.01)
RB_df['ReceivingShare'] = RB_df['Receiving_Tgt'] / (RB_df['Receiving_Tgt'] + RB_df['Rushing_Att']).replace(0, 0.01)
RB_df['PointsPerOpportunity'] = RB_df['FantasyPoints'] / (RB_df['Rushing_Att'] + RB_df['Receiving_Tgt']).replace(0, 0.01)

#workload filter
RB_df = RB_df[ (RB_df['Rushing_Att'] + RB_df['Receiving_Tgt']) >= 50 ]

#underrated score
RB_df['UnderratedScore'] = (
    RB_df['PointsPerOpportunity'] * 0.3 +
    RB_df['ReceivingShare'] * 0.2 +
    RB_df['YardsPerRush'] * 0.2 +
    RB_df['FantasyPointsPerGame'] * 0.2 +
    RB_df['Receiving_Rec'] * 0.1 -
    RB_df['FumblesTotal'] * 0.1
)


RB_df.sort_values('UnderratedScore', ascending=False).head(10)
```

```python
#wr subset
WR_df = df[df['Position'] == 'WR'].copy()

#fantasy ppg
WR_df['FantasyPointsPerGame'] = WR_df['FantasyPoints'] / WR_df['GamesPlayed']

#core metrics
WR_df['FumblesTotal'] = WR_df['Fumbles_Fmb'] + WR_df['Fumbles_FL']
WR_df['PointsPerOpportunity'] = WR_df['FantasyPoints'] / (WR_df['Receiving_Tgt'] + WR_df['Rushing_Att']).replace(0, 0.01)
WR_df['YardsPerTarget'] = WR_df['Receiving_Yds'] / WR_df['Receiving_Tgt'].replace(0, 0.01)


#filtering top players
WR_df = WR_df[WR_df['Fantasy_OvRank'] > 12]

#underrated score
WR_df['UnderratedScore'] = (
    WR_df['PointsPerOpportunity'] * 0.3 +
    WR_df['YardsPerTarget'] * 0.2 +
    WR_df['Receiving_TD'] * 0.15 +
    WR_df['FantasyPointsPerGame'] * 0.2 -
    WR_df['FumblesTotal'] * 0.1
)


WR_df.sort_values('UnderratedScore', ascending=False).head(10)
```

```python
#te subset
TE_df = df[df['Position'] == 'TE'].copy()

#fantasy ppg
TE_df['FantasyPointsPerGame'] = TE_df['FantasyPoints'] / TE_df['GamesPlayed']

#metrics
TE_df['FumblesTotal'] = TE_df['Fumbles_Fmb'] + TE_df['Fumbles_FL']
TE_df['CatchRate'] = TE_df['Receiving_Rec'] / TE_df['Receiving_Tgt'].replace(0, 0.01)
TE_df['YardsPerCatch'] = TE_df['Receiving_Yds'] / TE_df['Receiving_Rec'].replace(0, 0.01)
TE_df['PointsPerOpportunity'] = TE_df['FantasyPoints'] / (TE_df['Receiving_Rec'] + TE_df['Receiving_Tgt']).replace(0, 0.01)


#underrated score
TE_df['UnderratedScore'] = (
    TE_df['PointsPerOpportunity'] * 0.35 +
    TE_df['CatchRate'] * 0.25 +
    TE_df['YardsPerCatch'] * 0.15 +
    TE_df['FantasyPointsPerGame'] * 0.15 +
    TE_df['Receiving_Tgt'] * 0.1 -
    TE_df['FumblesTotal'] * 0.1
)


TE_df.sort_values('UnderratedScore', ascending=False).head(10)

```

```python
##Import API containing ADP data##
import requests

# API endpoint for 2024 PPR ADP, 12-team league
url = "https://fantasyfootballcalculator.com/api/v1/adp/ppr"
params = {
    "teams": 12,
    "year": 2024,
    "pos": "all"   # can also be "qb", "rb", "wr", "te"
}


response = requests.get(url, params=params)


if response.status_code == 200:
    data = response.json()
    adp_df = pd.DataFrame(data["players"])
    print(adp_df.head())
else:
    print("API call failed:", response.status_code)

#rename 'name' to 'Player'
adp_df = adp_df.rename(columns={"name": "Player", "adp": "ADP"})

#merge gor each position
merged_qb = pd.merge(QB_df, adp_df[['Player', 'ADP']], on='Player', how='inner')
merged_qb.head()

#RB merge
merged_rb = pd.merge(RB_df, adp_df[['Player', 'ADP']], on='Player', how='inner')
merged_rb.head()

#WR merge
merged_wr = pd.merge(WR_df, adp_df[['Player', 'ADP']], on='Player', how='inner')
merged_wr.head()


#TE merge
merged_te = pd.merge(TE_df, adp_df[['Player', 'ADP']], on='Player', how='inner')
merged_te.head()
```

```python
##2025 ADP predicition model##

#RB ADP prediction features
merged_rb["ADP_2024_Scaled"] = merged_rb["ADP"] * 0.05

adp_features = [
    "FantasyPoints", "FantasyPointsPerGame", "Rushing_Yds", "Rushing_TD",
    "Receiving_Yds", "Receiving_TD", "Receiving_Rec", "GamesPlayed","ADP_2024_Scaled",'PointsPerOpportunity','ReceivingShare','YardsPerRush','FumblesTotal'
]

X_adp = merged_rb[adp_features]
y_adp = merged_rb["ADP"]  # from 2024 fantasy draft

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

X_adp_train, X_adp_test, y_adp_train, y_adp_test = train_test_split(X_adp, y_adp, test_size=0.2, random_state=42)
adp_model = RandomForestRegressor(n_estimators=100, random_state=42)
adp_model.fit(X_adp_train, y_adp_train)


# Predict 2025 ADP
merged_rb["PredictedADP_2025"] = adp_model.predict(X_adp)
merged_rb["PredictedADP_DIFF"] = merged_rb["PredictedADP_2025"] - merged_rb["ADP"]
merged_rb.sort_values("PredictedADP_DIFF", ascending=True).head(10)
```

```python
#WR ADP prediction features
merged_wr["ADP_2024_Scaled"] = merged_wr["ADP"] * 0.05

adp_features = [
    "FantasyPoints", "FantasyPointsPerGame",
    "Receiving_Yds", "Receiving_TD", "Receiving_Rec", "GamesPlayed","ADP_2024_Scaled", 'FumblesTotal', 'PointsPerOpportunity','YardsPerTarget'
]

X_adp = merged_wr[adp_features]
y_adp = merged_wr["ADP"]  # from 2024 fantasy draft

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

X_adp_train, X_adp_test, y_adp_train, y_adp_test = train_test_split(X_adp, y_adp, test_size=0.2, random_state=42)
adp_model = RandomForestRegressor(n_estimators=100, random_state=42)
adp_model.fit(X_adp_train, y_adp_train)


# Predict 2025 ADP
merged_wr["PredictedADP_2025"] = adp_model.predict(X_adp)
merged_wr["PredictedADP_DIFF"] = merged_wr["PredictedADP_2025"] - merged_wr["ADP"]
merged_wr.sort_values("PredictedADP_DIFF", ascending=True).head(10)
```
